```python
from itertools import islice


def count(n):
    while True:
        yield n
        n += 1


"""

>>> c = count(0)
>>> c[10:20]
Traceback (most recent call last):
    ...
TypeError: 'generator' object is not subscriptable
>>> for x in islice(c, 10, 20):  # doctest: +ELLIPSIS
...     print(x)
10
11
12
...
17
18
19
"""

doctest.testmod()
```
